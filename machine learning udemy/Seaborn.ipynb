```python
import seaborn as sns
```

```python
#to see visualization inside the notebook
%matplotlib inline
```

```python
df=sns.load_dataset('tips')       #tips is built-in dataset
```

```python
df
```

```python
df.head()
```

```python
sns.distplot(df['total_bill'])   #this plot allows us to show distribution of univariance(one variable) set of observations
```

```python
#above shown is histogram  , the line is KDE(kernel density estimation)
```

```python
#to remove KDE
sns.distplot(df['total_bill'],kde=False)  
```

```python
#to change no. of bins
sns.distplot(df['total_bill'],kde=False,bins=30)
```

```python
#jointplot matches 2 distplots for bivariate data
sns.jointplot(x='total_bill',y='tip',data=df)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
 sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
```

```python
sns.pairplot(df)
```

```python
 sns.pairplot(df,hue='sex')
```

```python
#palette argument allows us to specify color for parameters(of hue column)
sns.pairplot(df,hue="sex",palette="coolwarm")
```

```python
sns.rugplot(df['total_bill'])
```

```python
#plotting KDE only
sns.kdeplot(df['total_bill'])
```
